```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time # for sleep time
import csv
import re
import functools
import datetime #for csv file title

driver = webdriver.Chrome()

driver.get('https://www.airbnb.com/s/Paris--France/homes?refinement_paths%5B%5D=%2Fhomes&place_id=ChIJD7fiBh9u5kcRYJSMaMOCCwQ&source=mc_search_bar&search_type=section_navigation&screen_size=large&checkin=2020-07-01&checkout=2020-07-02')

now = datetime.datetime.now() #for csv file title

#For csv file names
def reduce_concat(x, sep=""):
    return functools.reduce(lambda x, y: str(x) + sep + str(y), x)

def paste(*lists, sep=" ", collapse=None):
    result = map(lambda x: reduce_concat(x, sep=sep), zip(*lists))
    if collapse is not None:
        return reduce_concat(result, sep=collapse)
    return list(result)

#Open CSV Writer
filename = paste(['airBnb_Paris'], [now.year], ['-'], [now.month], ['-'], [now.day],['.csv'], sep='')
filename = ''.join(filename)
csv_file = open(filename, 'w', encoding='utf-8',newline = '')
writer = csv.writer(csv_file)

index = 1

house_all_links = []

while True:
    
    if index > 17:
        break
    try:
        print("Getting list of links from Page " + str(index))
        index = index + 1
        #driver.maximize_window()
        time.sleep(6) # give new page time to load
                
        house_list = (driver.find_elements_by_xpath('//a[@class="_i24ijs"]')) # find XPath here

        house_links = [house.get_attribute('href') for house in house_list ] 
        
        house_all_links.extend(house_links)
  
        # Locate the next button on the page.
        wait_button = WebDriverWait(driver, 10)
        
        next_button = wait_button.until(EC.element_to_be_clickable((By.XPATH,'//li[@class="_i66xk8d"]'))) # find XPath here
        
        next_button.click()

        
    except Exception as e:
        print(e)
        continue
        
print(house_all_links) ###***done for the list of all links***

```

```python
i=0

for url in house_all_links:
    i = i+1 #count
    print(i)
    # Initialize an empty dictionary for each review
    airbnb_dict = {}
    driver.get(url)

    print(url)
    
    try:
        
        time.sleep(3)
        
        ### getting house title
        try:
            house_title = driver.find_element_by_xpath('//h1/span[@class="_18hrqvin"]').text
        except:
            print("issue in house title")
            house_title = None
         
        ### getting house type
        try:
            house_type =  driver.find_element_by_xpath('//span[@class="_1p3joamp"]').text 
        except:
            print("issue in house_type1")
            house_type = None
            
        try:
            house_type2 =  driver.find_element_by_xpath('//div[@class="_hgs47m"]/div[@class="_n5lh69r"]/div[@class="_1p3joamp"]').text ### in case new layout
        except:
            print("issue in house_type2")
            house_type2 = None
 
        time.sleep(3)    

        ### getting price of the location
        try:
            price = driver.find_element_by_xpath('//span/span[@class="_doc79r"]').text
        except:
            print("issue in price")
            price = None        

        ### getting user name
        try:
            user = driver.find_element_by_xpath('//div[@class="_8b6uza1"]').text
        except:
            print("issue in user")
            user = None  

        ### getting total number of reviews
        try:
            no_of_ratings = driver.find_element_by_xpath('//span[@class="_1plk0jz1"]').text
        except:
            print("issue in no_of_ratings")
            no_of_ratings = None  
        
        time.sleep(3.5)
        
        ### getting rating - scale of 5
        try:
            rating = driver.find_element_by_xpath('//div[@class="_tghtxy2"]').text
        except:
            print("issue in rating")
            rating = None

        time.sleep(2)        
        
        ### getting house details - number of guests, bedroom, bath...
        try:
            max_occupancy = driver.find_elements_by_xpath('//div/div[@class="_36rlri"]/div[@class="_czm8crp"]')

            house_detail = [detail.text for detail in max_occupancy ]
        except:
            print("Issue in occupancy")
            house_detail = None

            
        ### getting all the information together     
        airbnb_dict['house_title'] = house_title
        airbnb_dict['house_type'] = house_type
        airbnb_dict['house_type2'] = house_type2
        airbnb_dict['price'] = price
        airbnb_dict['user'] = user
        airbnb_dict['no_of_reviews'] = no_of_ratings
        airbnb_dict['rating'] = rating
        airbnb_dict['house_detail'] = house_detail

        ### writing data to follow
        print("writing...")
        writer.writerow(airbnb_dict.values())

        print(house_title)
        print(house_type)
        print(house_type2)
        print(price)
        print(user)
        print(no_of_ratings)
        print(rating)
        print(house_detail)
        
        
        time.sleep(2)  

    except Exception as e:
        print(e)
        continue
        
csv_file.close()
driver.close()
```
